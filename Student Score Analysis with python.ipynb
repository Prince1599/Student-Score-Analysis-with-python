## Gender distribution

```python
plt.figure(figsize = (6,6))
ax = sns.countplot(data = df ,x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show
```

### From the above chart we have analysed that:
### The number of females in the data is more than the number of males

```python
gb = df.groupby("ParentEduc").agg({'MathScore' : 'mean' ,'ReadingScore':'mean','WritingScore':'mean'})
print(gb)
```

```python

plt.figure(figsize = (6,6))
sns.heatmap(gb,annot = True)
plt.title(" Relationship Between Parent Education and Student's Score")
plt.show()
```

### From the above chart we have concluded that the education of the parents have a great impact on their Score

```python
# we also check martialStatus impact on their student 
gb1 = df.groupby("ParentMaritalStatus").agg({'MathScore' : 'mean' ,'ReadingScore':'mean','WritingScore':'mean'})
print(gb1)
```

```python
plt.figure(figsize = (6,6))
sns.heatmap(gb1,annot = True)
plt.title("Relationship Between Parent Marital Status and Student's Score")
plt.show()
```

### From the above chart we have concluded that the MartialStatus of the Parent have negliable impact on student Score

## Distribution Of Ethnic Groups

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB= df.loc[(df["EthnicGroup"] == "group B")].count()
groupC= df.loc[(df["EthnicGroup"] == "group C")].count()
groupD= df.loc[(df["EthnicGroup"] == "group D")].count()
groupE= df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["group A","group B","group C","group D","group E"]
mylist = [groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
plt.pie(mylist,labels = l,autopct = "%1.2f%%")
plt.title(" Distribution of Ethnic Group")
plt.show()
```
